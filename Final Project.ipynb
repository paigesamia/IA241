Covid-19 began in March 2020 and has affected countries all around the world, including Lebanon. Covid-19 is an airborne spread disease that affects all ages.

```python
%matplotlib inline
import pandas
```

```python
df=pandas.read_excel('s3://samia-ia241-spring2021/COVID-19-geographic-disbtribution-worldwide-2020-12-14.xls')
df[:5]
```

## Question 1: How many cases have been confirmed in Lebanon?

```python
Lebanon_data=df.loc[df['countriesAndTerritories']=='Lebanon']
Lebanon_data[:10]
```

```python
Lebanon_cases=Lebanon_data.groupby('dateRep').sum()['cases']
Lebanon_cases.plot()
```

In January 2020 to July 2020, about no cases were reported. In September of 2020, cases began to spike at a fast rate with about 1000 confirmed cases. About 2000 positive cases in November 2020 were reported. Looking at the line chart, cases continued to increase, reaching to at least 3000 positive cases in Lebanon.

## Question 2: How Many Covid Related Deaths are there in Lebanon?

```python
Lebanon_deaths=Lebanon_data.groupby('dateRep').sum()['deaths']
Lebanon_deaths.plot()
```

In this line chart above, there has been a maximum of 25 covid related deaths. Between the months of March 2020 and September 2020, the number of reported deaths were under 10, but as of November 2020, Lebanon experienced a high spike of about 25 deaths.

## Question 3: How are the cases in Lebanon compared to other countries across Europe?

```python
Europe_data=df.loc[df['continentExp']=='Europe']
Europe_data[:10]
```

Compared to Albania, Lebanon and Albania have experienced about the same number of Covid deaths for the month of December. Although Lebanon reached a maximum of 25 deaths, both countries share similarities in the incline of covid-19 deaths.
